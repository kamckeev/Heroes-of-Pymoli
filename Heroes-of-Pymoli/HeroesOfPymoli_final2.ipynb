```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Create bins in which data will be held. Bins are <10, 10-14, 15-19, 20-24, 25-29, 30-34, 35-39 >39.
name_age= purchase_data[["Age","SN"]]
name_age_unique=name_age.drop_duplicates()
#print(name_age_unique)
age_bins= [0,9,14,19,24,29,34,39,100]
labels=["Under 10","10 to 14", "15 to 19", "20 to 24", "25 to 29", "30 to 34", "35 to 39", "Over 40"]
Age_groups = pd.cut(name_age_unique["Age"], age_bins, labels=labels)
Birthdays=Age_groups.value_counts()
Birthday_percent=((Birthdays/player_count)*100).round(2)
grouped_ages=pd.DataFrame({"Total Count": Birthdays, "Player %": Birthday_percent})
grouped_ages.sort_index()
```

## Purchasing Analysis (Age)

```python
# Bin the purchase_data data frame by age
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], age_bins, labels=labels)

# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
age_purchase_total = purchase_data.groupby(["Age Ranges"]).sum()["Price"].rename("Total Purchase Value")
age_average_total = purchase_data.groupby(["Age Ranges"]).mean()["Price"].rename("Average Purchase Price")
age_counts = purchase_data.groupby(["Age Ranges"]).count()["Price"].rename("Purchase Count")
age_average_purchase=age_purchase_total/Birthdays
#normalized_total = age_purchase_total / age_demographics["Total Count"]
# Create a summary data frame to hold the results
age_data = pd.DataFrame({"Purchase Count": age_counts, "Total Purchase Value": age_purchase_total, 
                         "Average Purchase Price": age_average_total, "Average Total Purchased": (age_average_purchase).round(2)})

# Optional: give the displayed data cleaner formatting
age_data["Average Purchase Price"] = age_data["Average Purchase Price"].map("${:,.2f}".format)
age_data["Total Purchase Value"] = age_data["Total Purchase Value"].map("${:,.2f}".format)
age_data ["Purchase Count"] = age_data["Purchase Count"].map("{:,}".format)
age_data["Average Total Purchased"]=age_data["Average Total Purchased"].map("${:,.2f}".format)
age_data = age_data.loc[:, ["Purchase Count", "Total Purchase Value","Average Purchase Price", "Average Total Purchased"]]

# Display the summary data frame
age_data

```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Basic Calculations
user_total = purchase_data.groupby(["SN"]).sum()["Price"]
user_average = purchase_data.groupby(["SN"]).mean()["Price"]
user_count = purchase_data.groupby(["SN"]).count()["Price"]

# Convert to DataFrame
user_data = pd.DataFrame({"Total Purchase Value": user_total, "Average Purchase Price": user_average, "Purchase Count": user_count})
user_data =user_data.sort_values("Total Purchase Value", ascending=False)

# Minor Data Munging
user_data["Average Purchase Price"] = user_data["Average Purchase Price"].map("${:,.2f}".format)
user_data["Total Purchase Value"] = user_data["Total Purchase Value"].map("${:,.2f}".format)
user_data = user_data.loc[:,["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]
user_data.head()

# Display Table
#user_data.sort_values("Total Purchase Value",ascending=False).head()
```

```python
#Most Popular Items
 
#Retrieve the Item ID, Item Name, and Item Price columns
items_bycount = purchase_data.groupby(['Item ID','Item Name'])['Price'].agg(['count','sum','mean']).sort_values(['count'], ascending= False).reset_index()


top_items = items_bycount.rename(columns = {'count': 'Purchase Count', 'mean' : 'Item Price',
                                            'sum' : 'Total Purchase Value' })


#Optional: give the displayed data cleaner formatting
top_items_df = top_items.set_index(['Item ID', 'Item Name'])
top_items_df=top_items_df[['Purchase Count', 'Item Price', 'Total Purchase Value']]
top_items_df.head()

#Display a preview of the summary data frame
```
